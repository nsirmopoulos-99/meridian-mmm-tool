## Step 3 — Set up ngrok (creates your public URL)

## Step 4 — Start the backend server

```
import nest_asyncio
import uvicorn
import threading
from pyngrok import ngrok

nest_asyncio.apply()

# Open a public tunnel to port 8000
public_url = ngrok.connect(8000)

print('=' * 60)
print('🚀 BACKEND IS RUNNING!')
print('=' * 60)
print(f'\n🌐 YOUR PUBLIC URL:')
print(f'   {public_url.public_url}')
print(f'\n📋 PASTE THIS URL into the frontend tool')
print(f'\n⚠️  Keep this Colab tab open while running analysis')
print('=' * 60)

# Start FastAPI server
uvicorn.run(
    'app:app',
    host='0.0.0.0',
    port=8000,
    log_level='info'
)
```
